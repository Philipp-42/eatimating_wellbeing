### Creating a Artificial Dataset
First Starting Point generate some random data, from two underlying Factors (normal distributions).

```python
import numpy as np
import pandas as pd

np.random.seed(42)

num_samples = 1000

data_dist1 = np.random.normal(loc=0, scale=1, size=num_samples)
data_dist2 = np.random.normal(loc=5, scale=2, size=num_samples)

observed_series = pd.DataFrame()
```

```python
for i in range(3):
    observed_series[f'Series_{i+1}_Dist1_Transform1'] = data_dist1 + np.random.normal(loc=0, scale=0.5, size=num_samples)
    observed_series[f'Series_{i+1}_Dist2_Transform1'] = data_dist2 + np.random.normal(loc=0, scale=0.5, size=num_samples)

    observed_series[f'Series_{i+1}_Dist1_Transform2'] = 2 * data_dist1 + np.random.normal(loc=0, scale=1, size=num_samples)
    observed_series[f'Series_{i+1}_Dist2_Transform2'] = 2 * data_dist2 + np.random.normal(loc=0, scale=1, size=num_samples)

    observed_series[f'Series_{i+1}_Dist1_Transform3'] = np.sin(data_dist1) + np.random.normal(loc=0, scale=0.2, size=num_samples)
    observed_series[f'Series_{i+1}_Dist2_Transform3'] = np.sin(data_dist2) + np.random.normal(loc=0, scale=0.2, size=num_samples)

    observed_series[f'Series_{i+1}_Dist1_Transform4'] = data_dist1**2 + np.random.normal(loc=0, scale=0.5, size=num_samples)
    observed_series[f'Series_{i+1}_Dist2_Transform4'] = data_dist2**2 + np.random.normal(loc=0, scale=0.5, size=num_samples)
    
    observed_series[f'Series_{i+1}_Dist1_Transform5'] = np.exp(data_dist1) + np.random.normal(loc=0, scale=0.5, size=num_samples)
    observed_series[f'Series_{i+1}_Dist2_Transform5'] = np.exp(data_dist2) + np.random.normal(loc=0, scale=0.5, size=num_samples)

    observed_series[f'Series_{i+1}_Dist1_Transform6'] = np.cos(data_dist1) + np.random.normal(loc=0, scale=0.2, size=num_samples)
    observed_series[f'Series_{i+1}_Dist2_Transform6'] = np.cos(data_dist2) + np.random.normal(loc=0, scale=0.2, size=num_samples)
```
